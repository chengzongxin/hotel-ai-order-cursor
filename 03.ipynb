```python
from dotenv import load_dotenv
load_dotenv(override=True)

from repositories.product_store import get_product_store

store = get_product_store()
print("向量库加载完成")
```

```python
# 测试五类场景的检索效果
# query = 用户说的商品 + 故障现象（安装/测量无故障则只有商品名）

test_queries = [
    ("空调 不制冷",   "单次维修服务"),
    ("水龙头 漏水",   "单次维修服务"),
    ("洗衣机",        "单次安装"),
    ("窗帘",          "单次测量"),
    ("门锁 打不开",   "单次维修服务 / 托管维修"),
]

for query, expected_type in test_queries:
    print(f"\n{'='*55}")
    print(f"query: 「{query}」  期望类型: {expected_type}")
    results = store.search(query, top_k=3, threshold=0.4)
    if not results:
        print("  无匹配结果（低于阈值）")
    for r in results:
        print(f"  [{r['score']:.3f}] {r['service_product_name']!r}  →  {r['service_order_type']}")
```
